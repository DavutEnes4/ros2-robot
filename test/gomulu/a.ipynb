# Ders 1 Ders Notundaki örnekler

```python
import networkx as nx
import matplotlib.pyplot as plt

# Ağ oluştur
G = nx.Graph()
G.add_edges_from([
    ('Ali', 'Ayşe'),
    ('Ali', 'Mehmet'),
    ('Ayşe', 'Zeynep'),
    ('Mehmet', 'Zeynep'),
    ('Zeynep', 'Can')
])

# Ağ çizimi
plt.figure(figsize=(6, 4))
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000)
plt.title("Basit Sosyal Ağ")
plt.show()
```

```python
G2 = nx.Graph()
G2.add_edges_from([
    ("Mine","Ravza"),
    ("Ravza","Enes"),
    ("Mine","Enes"),
    ("Enes","Taha"),
    ("Taha","Davut"),
    ("Enes","Davut"),
    ("Mücahit","Mine"),
    ("Mücahit","Ravza"),
    ("Mücahit","Taha"),
    ("Mücahit","Davut"),
    ("Gülsüm","Mine"),
    ("Gülsüm","Ravza"),
    ("Gülsüm","Taha"),
    ("Gülsüm","Davut")
])

plt.figure(figsize=(6,4))
nx.draw(G2, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000)
plt.title("Basit Sosyal Ağ")
plt.show()
```

```python
# Derece merkeziyeti
degree_centrality = nx.degree_centrality(G)
print("Derece Merkeziyeti:", degree_centrality)

merkez_derecesi = nx.degree_centrality(G2)
print("Derece Merkeziyeti 2:", merkez_derecesi)
# PageRank (Yönlü ağlar için daha uygundur, ama örnek amaçlı kullanılabilir)
#pagerank = nx.pagerank(G2)
#print("PageRank:", pagerank)
```

```python
from networkx.algorithms.community import girvan_newman

# Toplulukları bul
communities = next(girvan_newman(G))
for i, community in enumerate(communities):
    print(f"Topluluk {i + 1}: {community}")

# Toplulukları bul
topluluk = next(girvan_newman(G2))
for i, uye in enumerate(topluluk):
    print(f"Topluluk {i+1}: {uye}")
```

```python
def derece_fonksiyon(G):
    degree_centrality = nx.degree_centrality(G)
    print("Derece Merkeziyeti:", degree_centrality)
    
def topluluk_fonksiyon(G):
    communities = next(girvan_newman(G))
    for i, community in enumerate(communities):
        print(f"Topluluk {i + 1}: {community}")
```

```python
H = nx.Graph()
H.add_edges_from([
    ('Leonardo', 'Kate'),
    ('Leonardo', 'Tom'),
    ('Tom', 'Kate'),
    ('Morgan', 'Brad'),
    ('Brad', 'Angelina')
])

plt.figure(figsize=(6,4))

nx.draw(H, with_labels=True, node_color='salmon', edge_color='black')
plt.title("Yardımcı Oyuncu Ağı")
plt.show()
```

```python
import random

# 10 düğüm: 5 kırmızı, 5 mavi
nodes = [(i, {'color': 'red' if i < 5 else 'blue'}) for i in range(10)]
HG = nx.Graph()
HG.add_nodes_from(nodes)

# Benzer düğümler arasında daha fazla bağlantı kur
for i in range(10):
    for j in range(i+1, 10):
        if HG.nodes[i]['color'] == HG.nodes[j]['color'] and random.random() < 0.6:
            HG.add_edge(i, j)
        elif HG.nodes[i]['color'] != HG.nodes[j]['color'] and random.random() < 0.2:
            HG.add_edge(i, j)

colors = [HG.nodes[i]['color'] for i in HG.nodes()]
plt.figure(figsize=(6, 4))
nx.draw(HG, with_labels=True, node_color=colors, edge_color='gray', node_size=1000)
plt.title("Homofili Etkisi")
plt.show()
```

```python
derece_fonksiyon(HG)
topluluk_fonksiyon(HG)
```

# Ders 2 NetworX de ki kodlar ve açıklamaları

```python
###
# Network kütüphanesi içeri aktaralım
# Yönsüz bir ağ oluşturalım ve içine iki adet düğüm ve 1 kenar ekleyelim 
###
import networkx as nx

Graph = nx.Graph() # Yönsüz ağ sınıfından bir nesne türetelim
Graph.add_node(1) # 1 isimli/ID li bir düğüm ekle
Graph.add_node(2) # 2 isimli/ID li bir düğüm ekle 
Graph.add_edge(1,2) # 1 ve 2 isimli düğümleri bağladık/Kenar ekledik.
```

```python
# Düğüm listesi
print(Graph.nodes())
# Kenar listesi
print(Graph.edges())
print(Graph.number_of_nodes())
print(Graph.number_of_edges())
# 2. düğümünün komşuları
Graph.neighbors(2)

# Olmayan bir düğümün komşularını yazdırmaya çalışırsak "NetworkXError" hatasını verir
# print(Graph.neighbors(3))
```

```python
###
# Network kütüphanesi içeri aktaralım
# Yönlü bir ağ oluşturalım ve içine bir kaç adet düğüm ve kenar ekleyelim 
###
import networkx as nx

DiGraph = nx.DiGraph() # Yönsüz ağ sınıfından bir nesne türetelim
DiGraph.add_edge(1,2) # 1 ve 2 ID'li düğüm oluşturur ve 1 den 2 ye yönlü bağlantı oluşturur.
DiGraph.add_edge(2,1) # 2 ve 1 ID'li düğümleri 2 den 1 e doğru yönlü bağlar

# Birden fazla yönlü kenar eklemek için
DiGraph.add_edges_from([(2,3),(3,4),(4,2)])
```

```python
print(DiGraph.number_of_nodes())
print(DiGraph.number_of_edges())
print(DiGraph.neighbors(2))
print(DiGraph.predecessors(2))
print(DiGraph.successors(2))
```

```python
# Ağ örneği üretmek için aşağıdaki komıtlar kullamılabilir.

B = nx.complete_bipartite_graph(4,5) # Bir biriyle bağlı iki farklı sıralı düğüm oluşturur.
C = nx.cycle_graph(4) # Bir birini sırasıyla takip eden halka şeklinde ağ üretir.
S = nx.star_graph(6) # Yıldız metolojisinde bir ağ örneği oluşturur.
P = nx.path_graph(5) # Düz bir bağlantı hattı oluşturur. 
```

## Yoğunluk ve Seyreklik

* Tam Ağ: Mevcut ıkab Tüm düğümler çift bağlıdır.

* Maksimum bağlantı sayısı L(Max)
* * Yönsüz L =  (N * (N - 1)) /2
* * Yönlü L =  N * (N - 1)

* Yoğunluk

```python
# Yönsüz G grafının yoğunluğu
yogunluk_yonsuz = nx.density(Graph)
print(yogunluk_yonsuz)

# Yönlü D grafının yoğunluğu
yogunluk_yonlu = nx.density(DiGraph)
print(yogunluk_yonlu)
```
