```python
import networkx as nx
import matplotlib.pyplot as plt

# AÄ oluÅtur
G = nx.Graph()
G.add_edges_from([
    ('Ali', 'AyÅe'),
    ('Ali', 'Mehmet'),
    ('AyÅe', 'Zeynep'),
    ('Mehmet', 'Zeynep'),
    ('Zeynep', 'Can')
])

# AÄ Ã§izimi
plt.figure(figsize=(6, 4))
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000)
plt.title("Basit Sosyal AÄ")
plt.show()
```

```python
G2 = nx.Graph()
G2.add_edges_from([
    ("Mine","Ravza"),
    ("Ravza","Enes"),
    ("Mine","Enes"),
    ("Enes","Taha"),
    ("Taha","Davut"),
    ("Enes","Davut"),
    ("MÃ¼cahit","Mine"),
    ("MÃ¼cahit","Ravza"),
    ("MÃ¼cahit","Taha"),
    ("MÃ¼cahit","Davut"),
    ("GÃ¼lsÃ¼m","Mine"),
    ("GÃ¼lsÃ¼m","Ravza"),
    ("GÃ¼lsÃ¼m","Taha"),
    ("GÃ¼lsÃ¼m","Davut")
])

plt.figure(figsize=(6,4))
nx.draw(G2, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000)
plt.title("Basit Sosyal AÄ")
plt.show()
```

```python
# Derece merkeziyeti
degree_centrality = nx.degree_centrality(G)
print("Derece Merkeziyeti:", degree_centrality)

merkez_derecesi = nx.degree_centrality(G2)
print("Derece Merkeziyeti 2:", merkez_derecesi)
# PageRank (YÃ¶nlÃ¼ aÄlar iÃ§in daha uygundur, ama Ã¶rnek amaÃ§lÄ± kullanÄ±labilir)
#pagerank = nx.pagerank(G2)
#print("PageRank:", pagerank)
```

```python
from networkx.algorithms.community import girvan_newman

# TopluluklarÄ± bul
communities = next(girvan_newman(G))
for i, community in enumerate(communities):
    print(f"Topluluk {i + 1}: {community}")

# TopluluklarÄ± bul
topluluk = next(girvan_newman(G2))
for i, uye in enumerate(topluluk):
    print(f"Topluluk {i+1}: {uye}")
```

```python
def derece_fonksiyon(G):
    degree_centrality = nx.degree_centrality(G)
    print("Derece Merkeziyeti:", degree_centrality)
    
def topluluk_fonksiyon(G):
    communities = next(girvan_newman(G))
    for i, community in enumerate(communities):
        print(f"Topluluk {i + 1}: {community}")
```

```python
import random

# 10 dÃ¼ÄÃ¼m: 5 kÄ±rmÄ±zÄ±, 5 mavi
nodes = [(i, {'color': 'red' if i < 5 else 'blue'}) for i in range(10)]
HG = nx.Graph()
HG.add_nodes_from(nodes)

# Benzer dÃ¼ÄÃ¼mler arasÄ±nda daha fazla baÄlantÄ± kur
for i in range(10):
    for j in range(i+1, 10):
        if HG.nodes[i]['color'] == HG.nodes[j]['color'] and random.random() < 0.6:
            HG.add_edge(i, j)
        elif HG.nodes[i]['color'] != HG.nodes[j]['color'] and random.random() < 0.2:
            HG.add_edge(i, j)

colors = [HG.nodes[i]['color'] for i in HG.nodes()]
plt.figure(figsize=(6, 4))
nx.draw(HG, with_labels=True, node_color=colors, edge_color='gray', node_size=1000)
plt.title("Homofili Etkisi")
plt.show()
```

```python
derece_fonksiyon(HG)
topluluk_fonksiyon(HG)
```

# Ders 2 NetworX de ki kodlar ve aÃ§Ä±klamalarÄ±

```python
###
# Network kÃ¼tÃ¼phanesi iÃ§eri aktaralÄ±m
# YÃ¶nsÃ¼z bir aÄ oluÅturalÄ±m ve iÃ§ine iki adet dÃ¼ÄÃ¼m ve 1 kenar ekleyelim 
###
import networkx as nx

Graph = nx.Graph() # YÃ¶nsÃ¼z aÄ sÄ±nÄ±fÄ±ndan bir nesne tÃ¼retelim
Graph.add_node(1) # 1 isimli/ID li bir dÃ¼ÄÃ¼m ekle
Graph.add_node(2) # 2 isimli/ID li bir dÃ¼ÄÃ¼m ekle 
Graph.add_edge(1,2) # 1 ve 2 isimli dÃ¼ÄÃ¼mleri baÄladÄ±k/Kenar ekledik.
```

```python
# DÃ¼ÄÃ¼m listesi
print(Graph.nodes())
# Kenar listesi
print(Graph.edges())
print(Graph.number_of_nodes())
print(Graph.number_of_edges())
# 2. dÃ¼ÄÃ¼mÃ¼nÃ¼n komÅularÄ±
Graph.neighbors(2)

# Olmayan bir dÃ¼ÄÃ¼mÃ¼n komÅularÄ±nÄ± yazdÄ±rmaya Ã§alÄ±ÅÄ±rsak "NetworkXError" hatasÄ±nÄ± verir
# print(Graph.neighbors(3))
```

```python
###
# Network kÃ¼tÃ¼phanesi iÃ§eri aktaralÄ±m
# YÃ¶nlÃ¼ bir aÄ oluÅturalÄ±m ve iÃ§ine bir kaÃ§ adet dÃ¼ÄÃ¼m ve kenar ekleyelim 
###
import networkx as nx

DiGraph = nx.DiGraph() # YÃ¶nsÃ¼z aÄ sÄ±nÄ±fÄ±ndan bir nesne tÃ¼retelim
DiGraph.add_edge(1,2) # 1 ve 2 ID'li dÃ¼ÄÃ¼m oluÅturur ve 1 den 2 ye yÃ¶nlÃ¼ baÄlantÄ± oluÅturur.
DiGraph.add_edge(2,1) # 2 ve 1 ID'li dÃ¼ÄÃ¼mleri 2 den 1 e doÄru yÃ¶nlÃ¼ baÄlar

# Birden fazla yÃ¶nlÃ¼ kenar eklemek iÃ§in
DiGraph.add_edges_from([(2,3),(3,4),(4,2)])
```

```python
print(DiGraph.number_of_nodes())
print(DiGraph.number_of_edges())
print(DiGraph.neighbors(2))
print(DiGraph.predecessors(2))
print(DiGraph.successors(2))
```
